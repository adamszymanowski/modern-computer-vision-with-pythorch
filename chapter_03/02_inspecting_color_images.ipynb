## Inspecting Color Images

1. Download a color image:
- I used this [public domain vegetable-aisle-in-grocery-store.jpg](https://www.photos-public-domain.com/wp-content/uploads/2018/11/https://www.photos-public-domain.com/wp-content/uploads/2018/11/vegetable-aisle-in-grocery-store.jpg) one

2. Import the relevant packagees and load the image:

```python
import cv2, matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread('./data/vegetable-aisle-in-grocery-store.jpg')

plt.imshow(img)
```

3. Crop the image and change format from BGR to RGB

```python
img = img[2450:2650, 500:600, :]
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

4. Plot the image, show its shape

```python
print(img.shape)
plt.imshow(img)
```

5. The bottom-right 3x3 array of pixels can be obtained as follows:

```python
crop = img[-3:,-3:]
```

6. Print and plot the pixel values:

```python
print(crop)
plt.imshow(crop)
```
