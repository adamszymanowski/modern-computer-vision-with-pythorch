# Preparing Our Data For Image Classification
We will work on a single dataset throughout this chapter - the Fashion MNIST dataset. Let's prepare this dataset:

1. Start by downloading the dataset and importing the relevant packages. The `torchvision` package contains various datasets - one of which is the `FashionMNIST` dataset, which we will be working on in this chapter:

```python
from torchvision import datasets
import torch
data_folder = './data/FMNIST'
fmnist = datasets.FashionMNIST(data_folder, download=True, train=True)
train_images = fmnist.data
train_labels = fmnist.targets 
```

2. Inspect the tensors that we are dealing with:

```python
unique_labels = train_labels.unique()
print(f"training images: X -> {train_images.shape}")
print(f"training labels: Y -> {train_labels.shape}")
print(f"unique label values: {unique_labels}")
print(f"unique label classes: {fmnist.classes}")
```

3. Plot a random sample of 10 images for all the 10 possible classes:

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

grid_size = (10, 10)
rows, cols = grid_size
fig, ax = plt.subplots(rows, cols, figsize=grid_size)
for label_class, plot_row in enumerate(ax):
    label_x_rows = np.where(train_labels == label_class)[0]
    for plot_cell in plot_row:
        plot_cell.grid(False)
        plot_cell.axis('off')
        idx = np.random.choice(label_x_rows)
        x, y = train_images[idx], train_labels[idx]
        plot_cell.imshow(x, cmap='gray')

plt.tight_layout()
```

Note that in the preceeding image, each row represents a sample of 10 different images beloning to the same class.
