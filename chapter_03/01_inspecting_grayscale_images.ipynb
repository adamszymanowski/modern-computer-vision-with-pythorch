# Representing an image

## Inspecting Grayscale images

1. Download an image
- I used this [public domain vegetable-aisle-in-grocery-store.jpg](https://www.photos-public-domain.com/wp-content/uploads/2018/11/https://www.photos-public-domain.com/wp-content/uploads/2018/11/vegetable-aisle-in-grocery-store.jpg) one

2. Import the `cv2` (OpenCV, for reading image from disk) and `matplotlib` (for plotting the loaded image) libraries and read the downloaded image into the Python environment: 

```python
%matplotlib inline
import cv2, matplotlib.pyplot as plt
img = cv2.imread('./data/vegetable-aisle-in-grocery-store.jpg')
```

3. Show image, and then convert it to grayscale

```python
print(img.shape)
plt.imshow(img)
plt.show() # used for showing multiple images
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap="gray")
plt.show()
```

4. Scale down image:

```python
small_height = img.shape[0]//40
small_width = img.shape[1]//40
print(small_height, small_width)

img_gray_small = cv2.resize(img_gray, (small_width, small_height))
plt.imshow(img_gray_small, cmap="gray")
```

5. Let's inspect the pixel values

```python
print(img_gray_small)
```
